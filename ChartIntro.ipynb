```python

countries = ["USA","China","Japan","GreatBritain","India","Australia","France","Canada" ]
goldMedalsEraened = [39,38,27,22,1,17,10,7]
total_medals = [113,88,58,65,71,46,40,24]
silver_medals = [41,32,14,21,28,7,12,12]

plt.title("TOKYO OLYMPICS 2021 Medal tally")
plt.xlabel("Contry Name")
plt.ylabel("Gold Medals Earned")
plt.bar(countries,goldMedalsEraened,color="green")

plt.show()
plt.savefig("graph.png")
```

```python
plt.plot(countries,goldMedalsEraened, 'r--', countries,total_medals,'g^')
```

```python
plt.plot(countries,silver_medals, 'bs') 
```

```python
plt.barh(countries,goldMedalsEraened,color="yellow")
```

```python
labels = "USA","China","Japan","GreatBritain","India","Australia","France","Canada"
fig1, ax1 = plt.subplots()

explode = (0, 0, 0.2, 0,0,0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')
ax1.pie(goldMedalsEraened, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sinc(x)

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
# Example of stacked bar chart


width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(countries, goldMedalsEraened, width, yerr=total_medals, label='goldMedals')
ax.bar(countries, silver_medals, width, yerr=total_medals, bottom=silver_medals,
       label='Silver Medals earned')

ax.set_ylabel('Medals Earned')
ax.set_title('Medals gold and silver')
ax.legend()

plt.show()
```
